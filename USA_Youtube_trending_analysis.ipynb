## Loading data

```python
df_videos = pd.read_csv('./data/USvideos.csv')
# df_videos
```

## Whats the hottest channel (Ordered by the top 10)? 

```python
df_channels = df_videos['channel_title']
counts = df_channels.value_counts()
counts.head(10).sort_values().plot(kind='barh')
```

## Whats the relations between 'likes' and 'views' per video?
