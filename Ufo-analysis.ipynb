### Basic analysis over more than 80 thousand register of UFOs events in whole world

## Classic lib importation

```python
# importing libraries
import pandas as pd
import matplotlib.pyplot as plt
% matplotlib inline
plt.style.use('ggplot')
```

## Cleaning some parts, getting up some questions

```python
df_ufos = pd.read_csv('./data/complete.csv', low_memory=False)
df_ufos = df_ufos.fillna('--')
```

## What's the most affected state?

```python
df_states = df_ufos['state']
df_states = df_states.dropna()
df_states.value_counts().head().plot(kind='pie', table=True)
```

## What's the most common formation of UFOs?

```python
df_shapes = df_ufos['shape']
df_shapes.fillna('--').value_counts().head(10).plot(kind='pie', table=True)
```

## How long it usually takes?

```python
df_duration = df_ufos['duration (seconds)']
df_duration = df_duration.value_counts()
df_duration = df_duration / 60 / 60 
df_duration.head(10).plot(kind='pie',table=True)
```
